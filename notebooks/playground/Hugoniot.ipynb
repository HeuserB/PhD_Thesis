```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

$H (p,v) = E(p,v)-E(p_0,v_0)+\frac {1}{2}(p + p_0)(v_0 - v)$ 

```python
sns.set_palette('dark')
fig, ax     =   plt.subplots(figsize=(8,6))

V0, alpha, dE       = 1., 1., 0.5
V                   = np.linspace(1.1,1.3,100)
P                   = -dE * 2. / (V0-V) 
sns.lineplot(x=V,y=P,label=f"H")
```
